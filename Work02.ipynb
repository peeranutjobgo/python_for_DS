```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display 
df_app = pd.read_csv('application.csv', index_col=None)
df_career = pd.read_csv('final_career.csv', index_col=None)
df_user = pd.read_csv('final_user.csv', index_col=None)
pd.options.display.max_columns = None
```

```python
#find missing value
print(df_user.userID.value_counts())
print('-----------------------------------------')
print(df_career.careerID.value_counts())
print('-----------------------------------------')
print(df_user.apply(lambda x: sum(x.isnull())))
print('-----------------------------------------')
print(df_career.apply(lambda x: sum(x.isnull())))
print('-----------------------------------------')
```

```python
df_career.drop_duplicates(subset ="careerID", keep = 'last', inplace = True) 
df_user.drop_duplicates(subset ="userID", keep = 'last', inplace = True) 
```

```python
df_user.gender = df_user.gender.replace(to_replace ="ช", value ="Man") 
df_user.gender = df_user.gender.replace(to_replace ="ญ", value ="Woman") 
df_user
```

```python
df_app2 = pd.merge(df_app, df_career, how="inner",on = ['careerID'])
df_app3 = pd.merge(df_app2, df_user, how="inner",on = ['userID'])
df_app3
```

```python
df_app3.userEduCategory = df_app3.userEduCategory.fillna('nomajorEduCategory')
df_app3.userLastWorkCategory = df_app3.userLastWorkCategory.fillna('noLastWork/newwork')
# df_app3.province = df_app3.province.fillna('unknow')
```
