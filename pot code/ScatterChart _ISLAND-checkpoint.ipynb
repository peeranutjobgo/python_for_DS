```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

nwdf = pd.read_csv('nwdf.csv')
```

```python
nwdf_is=nwdf[nwdf['ocean_proximity'] == 'ISLAND']
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_is['population'], nwdf_is['households'], fit_reg=True)
plt.title('Compared population with households in the ISLAND', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_is['households'], nwdf_is['median_income'], fit_reg=True)
plt.title('Compared households with median income in the ISLAND', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_is['households'], nwdf_is['median_house_value'], fit_reg=True)
plt.title('Compared households with median house value in the ISLAND', fontsize=18)
```
