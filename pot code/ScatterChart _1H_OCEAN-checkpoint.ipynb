```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

nwdf = pd.read_csv('nwdf.csv')
```

```python
nwdf_1h=nwdf[nwdf['ocean_proximity'] == '<1H OCEAN']
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_1h['population'], nwdf_1h['households'])
plt.title('Compared population with households in the <1H OCEAN', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_1h['households'], nwdf_1h['median_income'])
plt.title('Compared households with median income in the <1H OCEAN', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_1h['households'], nwdf_1h['median_house_value'])
plt.title('Compared households with median house value in the <1H OCEAN', fontsize=18)
```
