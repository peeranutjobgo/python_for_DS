```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

nwdf = pd.read_csv('nwdf.csv')
```

```python
group_nwdf=nwdf.groupby('ocean_proximity').mean().round(2) ; counts = nwdf.groupby(['ocean_proximity']).size() ; group_nwdf['count'] = counts
group_nwdf
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.barplot(group_nwdf.index, group_nwdf['count'], alpha=0.8)
plt.title('Amount of information ocean proximity', fontsize=18)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Ocean proximity', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.barplot(group_nwdf.index, group_nwdf['population'], alpha=0.8)
plt.title('Compare the mean of population in each ocean proximity', fontsize=18)
plt.ylabel('mean of population', fontsize=12)
plt.xlabel('Ocean proximity', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.barplot(group_nwdf.index, group_nwdf['households'], alpha=0.8)
plt.title('Compare the mean of households in each ocean proximity', fontsize=18)
plt.ylabel('mean of households', fontsize=12)
plt.xlabel('Ocean proximity', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.barplot(group_nwdf.index, group_nwdf['median_income'], alpha=0.8)
plt.title('Compare the mean of median income in each ocean proximity', fontsize=18)
plt.ylabel('mean of median income', fontsize=12)
plt.xlabel('Ocean proximity', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.barplot(group_nwdf.index, group_nwdf['median_house_value'], alpha=0.8)
plt.title('Compare the mean of median house value in each ocean proximity', fontsize=18)
plt.ylabel('mean of median house value', fontsize=12)
plt.xlabel('Ocean proximity', fontsize=12)
plt.show()
```
