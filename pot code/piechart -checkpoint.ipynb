```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

nwdf = pd.read_csv('nwdf.csv')
```

```python
group_nwdf=nwdf.groupby('ocean_proximity').mean().round(2) ; counts = nwdf.groupby(['ocean_proximity']).size() ; group_nwdf['count'] = counts
```

```python
labels = group_nwdf.index
sizes = group_nwdf['count']
explode = (0, 0, 0, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.2f%%', shadow=True, startangle=90)

plt.axis('equal')
plt.title('Piechart percent of ocean proximity', fontsize=18)
plt.tight_layout()
plt.show()
```
