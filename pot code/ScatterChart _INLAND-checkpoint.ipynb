```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

nwdf = pd.read_csv('nwdf.csv')
```

```python
nwdf_in=nwdf[nwdf['ocean_proximity'] == 'INLAND']
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_in['population'], nwdf_in['households'], fit_reg=True)
plt.title('Compared population with households in the INLAND', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_in['households'], nwdf_in['median_income'], fit_reg=True)
plt.title('Compared households with median income in the INLAND', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_in['households'], nwdf_in['median_house_value'], fit_reg=True)
plt.title('Compared households with median house value in the INLAND', fontsize=18)
```
