```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

nwdf = pd.read_csv('nwdf.csv')
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf['population'], nwdf['households'])
plt.title('Compared population with households in the overall', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf['households'], nwdf['median_income'])
plt.title('Compared households with median income in the overall', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf['households'], nwdf['median_house_value'])
plt.title('Compared households with median house value in the overall', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf['households'], nwdf['total_rooms'])
plt.title('Compared households with total rooms in the overall', fontsize=18)
```
