```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display 
df_app = pd.read_csv('application.csv', index_col=None)
df_career = pd.read_csv('final_career.csv', index_col=None)
df_user = pd.read_csv('final_user.csv', index_col=None)
pd.options.display.max_columns = None
```

<font size="3"> ลบข้อมูลซํ้าซ้อนใน df_user โดยเหลือไว้เฉพาะตัวล่าสุด <font>

```python
df_user.drop_duplicates(subset ="userID", keep = 'last', inplace = True) 
df_user.userID.value_counts()
```

<font size="3"> ลบข้อมูลซํ้าซ้อนใน df_career โดยเหลือไว้เฉพาะตัวล่าสุด <font>

```python
df_career.drop_duplicates(subset ="careerID", keep = 'last', inplace = True) 
df_career.careerID.value_counts()
```

<font size="3"> เติม missing value ใน df_user ดังนี้ <font>

```python
# เติม missing value ใน gender ด้วยค่าข้อมูลก่อนหน้า
df_user.gender.fillna(method='ffill', inplace=True)
# เติม missing value ใน age ด้วยค่า mean
df_user.age.replace(0.0, np.NaN, inplace=True) 
df_user.age.fillna(df_user.age.mean(), inplace=True)
# เติม missing value = 0  ใน userEduLevel ด้วย nomajorEduCategory                                    
df_user.userEduLevel.replace(to_replace ="0", value ="nomajorEduCategory", inplace=True) 
# เติม missing value ใน userEduCategory ด้วย Other
df_user.userEduCategory = df_user.userEduCategory.fillna('Other')                                     
df_user.userEduCategory.replace(to_replace ="0", value ="Other", inplace=True) 
# เติม missing value ใน userLastWorkCategory ด้วย noLastWork/newwork                                      
df_user.userLastWorkCategory = df_user.userLastWorkCategory.fillna('noLastWork/newwork')
df_user.userLastWorkCategory.replace(to_replace ="0", value ="noLastWork/newwork", inplace=True) 
#เติม missing value ใน province ด้วย unknow
df_user.province = df_user.province.str.replace('\d+', '')
df_user.province.replace(to_replace ='', value ="unknow", inplace=True) 
df_user.province.replace(to_replace ="0", value ="unknow", inplace=True) 
df_user.province.fillna('unknow', inplace=True) 
# เติม missing value ใน userLastWorkExpMonth ด้วยค่า mean
df_user.userLastWorkExpMonth.replace(to_replace =-4, value =df_user.userLastWorkExpMonth.mean(), inplace=True) 
df_user
```

<font size="3"> เติม missing value ใน df_career ดังนี้ <font>

```python
# เติม missing value ใน careerCategory ด้วย Other และเปลี่ยน Others เป็น Other
df_career.careerCategory = df_career.careerCategory.fillna('Other')
df_career.careerCategory.replace(to_replace ="None", value ="Other", inplace=True) 
df_career.careerCategory.replace(to_replace ="Others", value ="Other", inplace=True)
df_career
```

<font size="3"> ผสาน table df_app กับ df_career เข้าด้วยกันกลายเป็น df_app2 <font>

```python
df_app2 = pd.merge(df_app, df_career, how="left",on = ['careerID'])
df_app2.apply(lambda x: sum(x.isnull()))
```

<font size="3"> ผสาน table df_app2 กับ df_user เข้าด้วยกันกลายเป็น df_app3 <font>

```python
df_app3 = pd.merge(df_app2, df_user, how="left",on = ['userID'])
df_app3.apply(lambda x: sum(x.isnull()))
```

```python
careerCategory_g = df_app3.groupby('careerCategory').mean().round(2) ; counts = df_app3.groupby(['careerCategory']).size() ; careerCategory_g['count'] = counts
careerCategory_g = careerCategory_g.sort_values(by = 'count', ascending=False)
plt.figure(figsize=(50,30))
sns.set(font_scale=3.0)
ax = sns.barplot(careerCategory_g['count'], careerCategory_g.index, alpha=0.8, order = careerCategory_g.index)
ax.set(xscale="log")
plt.title('The frequency of career category', fontsize=80)
ax.set_ylabel('Career Category', fontsize=60)
ax.set_xlabel('Amount of applyID', fontsize=60)
plt.savefig('careerCategory.png')
plt.show()
```
