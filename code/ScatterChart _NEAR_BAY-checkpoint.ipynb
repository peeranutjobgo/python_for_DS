```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

nwdf = pd.read_csv('nwdf.csv')
```

```python
nwdf_nb=nwdf[nwdf['ocean_proximity'] == 'NEAR BAY']
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_nb['population'], nwdf_nb['households'], fit_reg=True)
plt.title('Compared population with households in the NEAR BAY', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_nb['households'], nwdf_nb['median_income'], fit_reg=True)
plt.title('Compared households with median income in the NEAR BAY', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_nb['households'], nwdf_nb['median_house_value'], fit_reg=True)
plt.title('Compared households with median house value in the NEAR BAY', fontsize=18)
```
