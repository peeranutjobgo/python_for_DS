```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

nwdf = pd.read_csv('nwdf.csv')
```

```python
nwdf_no=nwdf[nwdf['ocean_proximity'] == 'NEAR OCEAN']
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_no['population'], nwdf_no['households'], fit_reg=True)
plt.title('Compared population with households in the NEAR OCEAN', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_no['households'], nwdf_no['median_income'], fit_reg=True)
plt.title('Compared households with median income in the NEAR OCEAN', fontsize=18)
```

```python
plt.figure(figsize=(10,5))
sns.set(style="whitegrid")
sns.regplot(nwdf_no['households'], nwdf_no['median_house_value'], fit_reg=True)
plt.title('Compared households with median house value in the NEAR OCEAN', fontsize=18)
```
